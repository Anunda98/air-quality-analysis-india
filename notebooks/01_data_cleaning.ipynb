Importing libraries

```python
import pandas as pd
import numpy as np
```

Loading main dataset (city_day data)

```python
# 
df = pd.read_csv(r"C:\Users\A.K\DATA_SCIENCE_WORKSPACE\kaggle_projects\project_1_air_quality\data\city_day.csv")

# View shape
print("Shape:", df.shape)

# Preview data
df.head()
```

```python
#Data quality check
print("Missing values:\n", df.isnull().sum())
```

```python
#summary of dataset
print(df.info())
```

```python
# Keeping only useful columns for analysis
df_clean = df[[
    "City", "Date",
    "PM2.5", "PM10",
    "NO2", "CO", "SO2", "O3",
    "AQI", "AQI_Bucket"
]]
df_clean.head()
```

```python
# Converting "Date" to datetime
df_clean["Date"] = pd.to_datetime(df_clean["Date"])
df_clean.dtypes
```

```python
# Check missing values percentage
df_clean.isnull().mean() * 100
```

```python
#keep AQI rows
df_clean = df_clean.dropna(subset=["AQI"])
df_clean.head()
```

```python
#handle pollutants missing values by filling with median
pollutants = ["PM2.5", "PM10", "NO2", "CO", "SO2", "O3"]

for col in pollutants:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())
```

```python
#Sorting by time
df_clean = df_clean.sort_values(by=["City", "Date"])
```

```python
#final check
df_clean.isnull().sum()
```

```python
#the clean dataset is ready for analysis and modeling
df.head()
```

```python
#This saves your cleaned dataset
df_clean.to_csv("../data/cleaned_air_quality.csv", index=False)
```
